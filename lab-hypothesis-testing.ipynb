In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
from statsmodels.multivariate.manova import MANOVA
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
df.columns = df.columns.str.replace('.', '', regex=False)
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here
# H0 = los pókemon dragon y grass tienen la misma media de HP
# H1 = los pókemon dragon tienen en promedio más HP que los de grass
# p = 0.05
# if p<0.05 we reject H0
# if p>0.05 we fail to reject H0

# Filtrar los Pokémon de tipo Dragon y Grass
dragon_hp = df[df['type_1'] == 'Dragon']['hp']
grass_hp = df[df['type_1'] == 'Grass']['hp']

# Realizar la prueba T de dos muestras independientes (no asumiendo varianzas iguales)
t_stat, p_value = st.ttest_ind(dragon_hp, grass_hp, equal_var=False, alternative='greater')

# Resultados
print(f"Estadístico t: {t_stat:.4f}")
print(f"p-valor: {p_value:.4f}")
```

Los pokemon de tipo Dragon tienen en promedio más HP que los de tipo Grass

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#code here
# H0 = Legendary Pokemons have the same stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary
# H1 = Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary
# p = 0.05
# if p<0.05 we reject H0
# if p>0.05 we fail to reject H0

# Variables dependientes (estadísticas) que queremos comparar
stats_columns = ['hp', 'attack', 'defense', 'sp._atk', 'sp._def', 'speed']

# Variable independiente: si el Pokémon es Legendario o No-Legendario
df['legendary'] = df['legendary'].apply(lambda x: 'legendary' if x else 'non_legendary')

# Crear el modelo MANOVA
manova = MANOVA.from_formula('hp + attack + defense + sp_atk + sp_def + speed ~ legendary', data=df)

# Ejecutar la prueba MANOVA
manova_results = manova.mv_test()

# Mostrar los resultados
print(manova_results)
```

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# H0 = No hay diferencia significativa en los precios de las casas cercanas y lejanas a escuelas o hospitales.
# H1 = Las casas cercanas a escuelas o hospitales tienen precios más altos que las casas lejanas.
# p = 0.05
# if p<0.05 we reject H0
# if p>0.05 we fail to reject H0
```

```python
# Definir las coordenadas de la escuela y el hospital
school_coords = (-118, 37)
hospital_coords = (-122, 34)

# Función para calcular la distancia euclidiana
def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Calcular las distancias de cada casa a la escuela y al hospital
df['dist_to_school'] = df.apply(lambda row: euclidean_distance(row['longitude'], row['latitude'], *school_coords), axis=1)
df['dist_to_hospital'] = df.apply(lambda row: euclidean_distance(row['longitude'], row['latitude'], *hospital_coords), axis=1)

# Clasificar si una casa está cerca de una escuela o un hospital
df['close_to_school_or_hospital'] = (df['dist_to_school'] < 0.50) | (df['dist_to_hospital'] < 0.50)

# Dividir el conjunto de datos en casas cercanas y lejanas
close_houses = df[df['close_to_school_or_hospital']]['median_house_value']
far_houses = df[~df['close_to_school_or_hospital']]['median_house_value']

# Realizar la prueba T de dos muestras independientes (asumiendo varianzas desiguales)
t_stat, p_value = st.ttest_ind(close_houses, far_houses, equal_var=False, alternative='greater')

# Resultados
print(f"Estadístico t: {t_stat:.4f}")
print(f"p-valor: {p_value:.4f}")

# Conclusión
if p_value < 0.05:
    print("Rechazamos la hipótesis nula: Las casas cercanas a escuelas o hospitales son más caras en promedio.")
else:
    print("No rechazamos la hipótesis nula: No hay evidencia suficiente para decir que las casas cercanas a escuelas o hospitales son más caras.")
```
